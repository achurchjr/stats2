```python
import spacy
import en_core_web_lg

nlp = en_core_web_lg.load()

#Raw string 1
string1 = 'The dog is small'

#Raw string 2
string2 = 'The dog is big'

doc1=nlp(string1)
doc1
doc2=nlp(string2)
doc1.similarity(doc2)
```

```python
ex1 = nlp('wolf dog cat bird fish')
for token1 in ex1:
    for token2 in ex1:
        print((token1.text,token2.text), "Similarity = ",token1.similarity(token2))
```

```python
mylist = [(token1.text,token2.text, token1.similarity(token2)) for token2 in ex1 for token1 in ex1]

import pandas as pd

df = pd.DataFrame(mylist)

df.columns = ['Token1', 'Token2', 'Similarity']

df
```

```python
import spacy
from spacy.lang.en.stop_words import STOP_WORDS

document1 = 'The construction of roads and railways'
document2 = 'Roads and bridges project'

stopwords = list(STOP_WORDS)

nlp = en_core_web_lg.load()

docx = nlp(document1)

for token in docx:
    print(token.text)
    
word_frequencies = {}

for word in docx:
    if word.text not in stopwords:
        if word.text not in word_frequencies.keys():
            word_frequencies[word.text] = 1
        else:
            word_frequencies[word.text] += 1
            
word_frequencies
```
